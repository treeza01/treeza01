```python
from sklearn.impute import SimpleImputer
```

```python
imputer  = SimpleImputer(strategy = 'mean') 
```

```python
data = data.copy()
data['Median Household Income'] = imputer.fit_transform(data[['Median Household Income']]) 
```

```python
from matplotlib import pyplot as plt
```

```python
column_names = ['Zip Code', 'Price', 'Beds', 'Baths', 'Living Space', 
        'Zip Code Population', 'Zip Code Density',
       'Median Household Income', 'Latitude', 'Longitude']
```

```python
i=1
for col in column_names:
    plt.subplots_adjust(wspace=2, hspace=0.6)
    plt.subplot(3,4,i)
    i+=1
    plt.boxplot(data[col])
    plt.title(col)
```

```python
outlier_columns = ['Price', 'Beds', 'Baths', 'Living Space','Zip Code Population', 'Zip Code Density','Median Household Income']
```

```python
import numpy as np
```

```python
i=1
for col in outlier_columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3-Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    upper_index = np.where(data[col]  >= upper)[0]
    lower_index = np.where(data[col]  <= lower)[0]
    data.drop(index = upper_index, inplace = True)
    data.drop(index = lower_index, inplace = True)

    plt.subplots_adjust(wspace=0.6, hspace=0.6)
    plt.subplot(3,4,i)
    i+=1
    plt.boxplot(data[col])
    plt.title(col)

    data.reset_index(inplace= True)
    data.drop('index', axis=1, inplace = True)
```

```python
data.drop('Address',axis = 1,inplace=True)
```

```python
data.drop('State',axis = 1,inplace=True)
```

```python
data.drop('County',axis = 1,inplace=True)
```

```python
data.drop('City',axis = 1,inplace=True)
data
```

```python
x=data.drop('Price',axis=1)
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 1820)  
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.coef_   
```

```python
model.intercept_
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
pd.DataFrame({'True Values':y_test,'Predicted Values':y_pred})
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
